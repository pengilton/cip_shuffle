```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from os import path
```

We are checking how the threshold affects the runtime. We keep other parameters at default. We are ommiting the buffered_fy and use fy32.

```python
# A function which can sort dataframes in a list of a dataframes. Based on bubble sort.
def sort_dataframe(dataframes: list, key: str):
    swapped = True
    while (swapped):
        swapped = False
        for i in range(len(dataframes)-1):
            # dataframes[i] gives us the ith dataframe, dataframes[i][key] gives us the right column and 
            # dataframes[i][key][0] gives us the first element of that column. We can do this because 
            # key should be equal for the whole column.
            a = dataframes[i][key][0]
            b = dataframes[i+1][key][0]
            if a > b:
                temp = dataframes[i]
                dataframes[i] = dataframes[i+1]
                dataframes[i+1] = temp
                swapped = True
```

```python
# We load data
threshold_fy32_files = glob.glob(path.join("..", "cpp", "uni-machine", "v0.1.12", "threshold_fy32", "*.csv"))
dfs_threshold_fy32 = [pd.read_csv(f) for f in threshold_fy32_files]
```

```python
# We sort the data
sort_dataframe(dfs_threshold_fy32, "threshold")
```

```python
# We calculate runtime per integer
for df in dfs_threshold_fy32:
    df["runtime/integer [ns]"] = df["total_runtime"] / (df["total_runs"] * df["integers"])
```

```python
# We plot data
ax_xticks = [2 ** i for i in range(0, 34, 2)]
ax_yticks = [1 * i for i in range(0, 50, 1)]

ax = dfs_threshold_fy32[0].plot(x="integers", y="runtime/integer [ns]", label=dfs_threshold_fy32[0]["function"][0] + " " + str(dfs_threshold_fy32[0]["threshold"][0]))
for i in range(1, len(dfs_threshold_fy32)):
    dfs_threshold_fy32[i].plot(ax=ax, x="integers", y="runtime/integer [ns]", label=dfs_threshold_fy32[i]["function"][0] + " " + str(dfs_threshold_fy32[i]["threshold"][0]))

ax.set_xscale("log", base=2)
ax.set_yscale("log", base=10)
ax.set_xticks(ax_xticks)
# ax.set_yticks(ax_yticks)
ax.set_xlim(2**0, 2**33)
# ax.set_ylim(0, 16)
ax.grid(True)
ax.set_xlabel("integers")
ax.set_ylabel("runtime/integer [ns]")
ax.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
ax.set_title("Comapring cip_shuffle threshold with fy32")
```
